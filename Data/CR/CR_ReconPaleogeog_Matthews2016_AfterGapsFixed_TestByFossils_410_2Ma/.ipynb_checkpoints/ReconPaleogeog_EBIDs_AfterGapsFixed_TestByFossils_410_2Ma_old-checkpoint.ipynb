## ReconPaleogeog_EBIDs_BeforeGapsFixed_TestByFossils_410_2Ma

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#data
Data = pd.read_csv('/Users/wenchaocao/Research/5_Consistency_Ratios_Fossils_Paleogeography/ReconPaleogeog_EBIDs_BeforeGapsFixed_TestByFossils_410_2Ma/ReconPaleogeog_EBIDs_BeforeGapsFixed_TestByFossils_410_2Ma.csv', delimiter=',')  
#CR
```

```python
plt.figure(figsize=(12,8),tight_layout=True) #figsize=(12,8)

#build a new array for installing rows twice, like row1, row1, row2, row2...
ConsistencyRatio_Marine = np.zeros(len(Data['Marine Fossil Consistency Ratio'])*2)
ConsistencyRatio_Marine[::2] = 100*np.asarray(Data['Marine Fossil Consistency Ratio'])
ConsistencyRatio_Marine[1::2] = 100*np.asarray(Data['Marine Fossil Consistency Ratio'])

ConsistencyRatio_Terrestrial = np.zeros(len(Data['Terrestrial Fossil Consistency Ratio'])*2)
ConsistencyRatio_Terrestrial[::2] = 100*np.asarray(Data['Terrestrial Fossil Consistency Ratio'])
ConsistencyRatio_Terrestrial[1::2] = 100*np.asarray(Data['Terrestrial Fossil Consistency Ratio'])

#create an array to install FromAge and ToAge at the same time
TimeStep = np.zeros(len(Data['From_age'])*2)
TimeStep[::2] = Data['To_age']
TimeStep[1::2] = Data['From_age']
#print timestep
                         
plt.plot(TimeStep,ConsistencyRatio_Marine,'b--',linewidth=2.0)
plt.plot(TimeStep,ConsistencyRatio_Terrestrial,'r--',linewidth=2.0)

# x, y labels, ticks, scales, line width and so on
plt.xlabel('Time(Ma)', fontsize=20)
plt.xticks(np.hstack((Data['From_age'],Data['To_age'])), size=12, rotation=50) #, rotation=50
plt.yticks(size=12)
plt.ylabel('Consistency Ratio (%)', fontsize=20)
plt.xlim((402,2))
plt.ylim((20,100))
plt.grid()
#plt.setp(linewidth=1.5)
plt.legend(('Marine Fossils','Terrestrial Fossils'),bbox_to_anchor=(.8,.25), loc=2, borderaxespad=0.) #,bbox_to_anchor=(1.,1.), loc=2, borderaxespad=0.
plt.show
#plt.savefig("test.svg")

#plt.savefig('CR_2.ps',dpi=600)
#plt.savefig('CR_2.pdf',dpi=600)
```
