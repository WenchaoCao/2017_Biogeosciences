```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#data
DataFossils = pd.read_csv('/Users/wenchaocao/Research/5_Consistency_Ratios_Fossils_Paleogeography/CR_Golonka_Paleogeog_Paleobiology_410_0Ma/Consistency_Ratio_GolonkaPaleogeog.csv', delimiter=',')  
#CR
```

```python
CR_Terrestrial = 100*DataFossils['Consistent Terrestrial Fossils']/(DataFossils['Consistent Terrestrial Fossils'] + DataFossils['Inconsistent Terrestrial Fossils'])
CR_Terrestrial
```

```python
CR_Marine = 100*DataFossils['Consistent Marine Fossils']/(DataFossils['Consistent Marine Fossils'] + DataFossils['Inconsistent Marine Fossils'])
CR_Marine
```

```python
CR_Terrestrial = 100*DataFossils['Consistent Terrestrial Fossils']/(DataFossils['Consistent Terrestrial Fossils'] + DataFossils['Inconsistent Terrestrial Fossils'])
CR_Marine = 100*DataFossils['Consistent Marine Fossils']/(DataFossils['Consistent Marine Fossils'] + DataFossils['Inconsistent Marine Fossils'])

plt.figure(figsize=(12,8)) #figsize=(12,8)
plot = plt.plot(DataFossils['Time intervals'],CR_Terrestrial,'red',DataFossils['Time intervals'],CR_Marine,'blue')

# x, y labels, ticks, scales, line width and so on 
plt.xlabel('Time(Ma)', fontsize=16)
plt.xticks(np.hstack((DataFossils['Time intervals'][0],DataFossils['Time intervals'][1::2])), size=10) #, rotation=50
plt.ylabel('Consistency Ratio (%)', fontsize=16)
plt.xlim((402,0))
plt.ylim((0,100))
plt.grid()
plt.setp(plot, linewidth=1.5)
plt.legend(('Terrestrial Fossils','Marine Fossils'),bbox_to_anchor=(.35,.25), loc=2, borderaxespad=0.) #,bbox_to_anchor=(1.,1.), loc=2, borderaxespad=0.
plt.show
plt.savefig('CR_Golonka_Paleogeog.png',dpi=300)
```
