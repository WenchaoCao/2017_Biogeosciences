```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#data
CR = pd.read_csv('/Users/wenchaocao/Research/5_Consistency_Ratios_Fossils_Paleogeography/ReconPaleogeog_EBIDs_AfterGapsFixed_TestByFossils_410_2Ma/Consistency_Ratio_RotationCorrections_latest_Plot.csv', delimiter=',')  
CR
```

```python
plt.figure(figsize=(12,8),tight_layout=True) #figsize=(12,8)
plot = plt.plot(CR['Time intervals'],100*(CR['Consistent Marine Fossils']/(CR['Consistent Marine Fossils']+CR['Inconsistent Marine Fossils'])),'blue', CR['Time intervals'],100*(CR['Consistent Terrestrial Fossils']/(CR['Consistent Terrestrial Fossils']+CR['Inconsistent Terrestrial Fossils'])),'red')

# x, y labels, ticks, scales, line width and so on
#plt.xlabel('Time(Ma)', fontsize=20)
plt.xticks(np.hstack((CR['Time intervals'][0],CR['Time intervals'][1::2])), size=12, rotation=50, ) #, rotation=50
plt.ylabel('Consistency Ratio (%)', fontsize=20)
plt.xlim((402,2))
plt.ylim((20,100))
plt.grid()
#plt.setp(linewidth=1.5)
plt.legend(('Marine Fossils','Terrestrial Fossils'),bbox_to_anchor=(.8,.25), loc=2, borderaxespad=0.) #,bbox_to_anchor=(1.,1.), loc=2, borderaxespad=0.
plt.show
#plt.savefig("test.svg")

plt.savefig('CR_2.ps',dpi=600)
plt.savefig('CR_2.pdf',dpi=600)
```
