```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# import data
Fossil_data=np.loadtxt(fname='/Users/wenchaocao/Desktop/Amount_Fossils.csv', delimiter=',')  
# set x values
x = Fossil_data[:,0:1]
# set y values
y1 = Fossil_data[:,1:2] # Terrestrial fossils
y2 = Fossil_data[:,2:3] # Marine fossils
y3 = Fossil_data[:,3:4] # Both
# set the width of the bars
width = 8

# plot a bar diagram in which bars overlap each other
p1 = plt.bar(x, y3, width, color='y', alpha=0.5)  #edgecolor='b' or "none" can change the edge color of bars
p2 = plt.bar(x, y2, width, color='b', alpha=0.5)  #alpha: set transparency
p3 = plt.bar(x, y1, width, color='r', alpha=0.5)

# set x, y labels and legends
plt.xlabel('Time(Ma)')
plt.ylabel('Amount of collecitons')
plt.legend((p1[0], p2[0], p3[0]), ('Terrestrial', 'Marine', 'Both'))

# set x, y axies ticks
xticks = np.arange(0, 450, 50)
yticks = np.arange(0, 60001, 10000)
plt.xticks(xticks, weight='bold', size='large', rotation=50)
plt.yticks(np.arange(0, 60001, 10000))

plt.show()
```
