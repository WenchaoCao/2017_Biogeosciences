```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#data
Flooding_Ratio=pd.read_csv('/Users/wenchaocao/Research/8_Flooding_History_410_0Ma/Flooding_History_Python/Flooding_History_410_0Ma.csv', delimiter=',')  
#Flooding_Ratio
```

```python
plt.figure(figsize=(12,8)) #figsize=(12,8)
plot = plt.plot(Flooding_Ratio['Time_intervals'],Flooding_Ratio['Flooding_Ratio'],'blue')

# x, y labels, ticks, scales, line width and so on 
plt.xlabel('Time(Ma)', fontsize=16)
plt.xticks(np.hstack((Flooding_Ratio['Time_intervals'][0],Flooding_Ratio['Time_intervals'][1::2])), size=10) #, rotation=50
plt.ylabel('Flooding Ratio (%)', fontsize=16)
plt.xlim((402,0))
plt.grid()
#plt.ylim((0,40))
plt.setp(plot, linewidth=1.5)
#plt.legend()  #,bbox_to_anchor=(1.,1.), loc=2, borderaxespad=0.
plt.show
plt.savefig('Flooding_Ratio_3.png',dpi=300)
```
