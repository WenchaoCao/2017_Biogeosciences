```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#read data
Haq_2008 = pd.read_csv('/Users/wenchaocao/Research/8_FloodingHistory_410_2Ma/6_FloodingHistory_FromReconsPaleogeog_EBIDs_ModifiedByFossils_402-2Ma/Haq_Schutter_2008.csv', delimiter=',')  
Haq_1987 = pd.read_csv('/Users/wenchaocao/Research/8_FloodingHistory_410_2Ma/6_FloodingHistory_FromReconsPaleogeog_EBIDs_ModifiedByFossils_402-2Ma/Haq87_Longterm_Combined.csv', delimiter=',')
Dietmar_2008 = pd.read_csv('/Users/wenchaocao/Research/8_FloodingHistory_410_2Ma/6_FloodingHistory_FromReconsPaleogeog_EBIDs_ModifiedByFossils_402-2Ma/Sea_level_Dietmar_2008/Sea_level_Dietmar_2008.csv', delimiter=',')
Flooding_Ratio=pd.read_csv('/Users/wenchaocao/Research/8_FloodingHistory_410_2Ma/4_Areas_ReconsPaleogeog_EBIDs_AfterModifiedByFossils_402-2Ma/Areas_Paleogeog_Features_latest.csv', delimiter=',')  
Flooding_Ratio_vanderMeer=pd.read_excel('/Users/wenchaocao/Research/34_Paper_Biogeosciences/Papers/van_der_Meer_etal_2017_sea_level_strontium_isotopes_mmc1.xlsx', delimiter=',')  
Flooding_Ratio_vanderMeer.iloc[1:,0]
```

```python
#fig = plt.figure(figsize=(12,16)) #figsize=(12,8)
fig, ax1 = plt.subplots(figsize=(8,5))

ax1.plot(Flooding_Ratio['Time intervals'],Flooding_Ratio['R_Flooding (Ama/A_T-Land)']*100,'blue', linewidth=1.5)
ax1.set_xlabel('Time (Ma)', fontsize=14)
ax1.xaxis.set_label_position('top')
ax1.xaxis.set_ticks_position('top')
ax1.set_xlim((402,2))
ax1.set_xticks(np.hstack((Flooding_Ratio['Time intervals'][0],Flooding_Ratio['Time intervals'][1::2]))) #, size=12, rotation=50
ax1.xaxis.set_tick_params(labelsize=10,direction='out')
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=50)
ax1.set_ylabel('Flooding Ratio (%)', fontsize=12) #, color='b'
ax1.yaxis.set_tick_params(labelsize=12,direction='out')
ax1.grid()
#for tl in ax1.get_yticklabels():
#    tl.set_color('b')
#plt.legend()  #,bbox_to_anchor=(1.,1.), loc=2, borderaxespad=0.

ax2 = ax1.twinx()
ax2.plot(Haq_2008['Times'],Haq_2008['Sea_level'],'red', Haq_1987['Times'],Haq_1987['Sea_level'],'purple', Dietmar_2008['Times'],Dietmar_2008['Sea_level'],'green',linewidth=1.5)

ax2.set_ylabel('Sea level (m)', fontsize=12)
ax2.set_xlim((402,2))
ax2.yaxis.set_tick_params(labelsize=12,direction='out')

plt.legend(('Haq et al., 2008','Haq et al., 1987', 'This study', 'Muller et al., 2008'),bbox_to_anchor=(0.3,.95), loc=2, borderaxespad=0., fontsize=10) #,bbox_to_anchor=(1.,1.), loc=2, borderaxespad=0.

plt.show
#plt.savefig('SeaLevel_FloodingRatio_MakeLegend.pdf',dpi=600)
#plt.savefig('SeaLevel_FloodingRatio_MakeLegend.png',dpi=600)
#plt.savefig('SeaLevel_FloodingRatio_MakeLegend.ps',dpi=600)
```
