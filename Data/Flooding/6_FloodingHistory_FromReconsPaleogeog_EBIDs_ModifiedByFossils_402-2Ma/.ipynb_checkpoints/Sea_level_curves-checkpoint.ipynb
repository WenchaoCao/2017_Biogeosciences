```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#read data
Haq_2008 = pd.read_csv('/Users/wenchaocao/Desktop/Nico/Haq_Schutter_2008.csv', delimiter=',')  
Haq_2005 = pd.read_csv('/Users/wenchaocao/Research/8_Flooding_History_410_0Ma/Haq87_Longterm_Combined.csv', delimiter=',')
Dietmar_2008 = pd.read_csv('/Users/wenchaocao/Desktop/Nico/Sea_level_Dietmar_2008/Sea_level_Dietmar_2008.csv', delimiter=',')
#Dietmar_2008
```

```python
plt.figure(figsize=(12,4)) #figsize=(12,8)
plot = plt.plot(Haq_2008['Times'],Haq_2008['Sea_level'],'red',
                Haq_2005['Times'],Haq_2005['Sea_level'],'blue',
                Dietmar_2008['Times'],Dietmar_2008['Sea_level'],'orange')

# x, y labels, ticks, scales, line width and so on 
plt.xlabel('Time(Ma)', fontsize=16)
#plt.xticks()
#plt.xticks(np.hstack((Fossil_data['Time intervals'][0],Fossil_data['Time intervals'][1::2])), size=10) #, rotation=50
plt.ylabel('Sea level (m)', fontsize=16)
plt.xlim((550,0))
#plt.ylim((0,40))
plt.setp(plot, linewidth=1.5)
plt.legend(('Haq et al., 2008','Haq et al., 2005', 'Dietmar et al., 2008'),bbox_to_anchor=(.43,.95), loc=2, borderaxespad=0.) #,bbox_to_anchor=(1.,1.), loc=2, borderaxespad=0.
plt.show
plt.savefig('Sea_level_Haq.png',dpi=300)
```
